```python
from ultralytics import YOLO
import cv2
import numpy as np
from collections import defaultdict
import time
from google.colab.patches import cv2_imshow

# Load YOLOv10 model
model = YOLO("/content/drive/MyDrive/Yolov10_Objecttracking/yolov10m.pt")

# Load video
cap = cv2.VideoCapture("/content/drive/MyDrive/Yolov10_Objecttracking/inference/cow.mp4")

# Count line coordinates
count_line = {'x1': 190, 'y1': 412, 'x2': 966, 'y2': 412}
offset = 80

track_history = defaultdict(lambda: [])
tracked_ids = set()
cow_count = 0
start_time = time.time()

# Video writer to save output
width = int(cap.get(3))
height = int(cap.get(4))
out = cv2.VideoWriter('output_tracking.avi', cv2.VideoWriter_fourcc(*'XVID'), 20, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run object tracking
    results = model.track(
        source=frame,
        conf=0.25,
        persist=True,
        tracker="bytetrack.yaml",
        classes=19  # Class ID for cow in COCO
    )

    annotated = results[0].plot()
    boxes = results[0].boxes.xywh.cpu()
    ids = results[0].boxes.id
    classes = results[0].boxes.cls

    # Draw count line
    cv2.line(annotated, (count_line['x1'], count_line['y1']), (count_line['x2'], count_line['y2']), (0, 127, 255), 3)

    if ids is not None:
        for box, track_id, cls_id in zip(boxes, ids.cpu().tolist(), classes.cpu().tolist()):
            x, y, w, h = box
            center_y = y + h / 2

            if count_line['y1'] <= center_y <= count_line['y1'] + offset:
                if track_id not in tracked_ids:
                    tracked_ids.add(track_id)
                    cow_count += 1

    # Display count and FPS
    fps = round(1 / (time.time() - start_time), 2)
    start_time = time.time()
    cv2.putText(annotated, f"FPS: {fps}", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 2)
    cv2.putText(annotated, f"Cows: {cow_count}", (10, 80), cv2.FONT_HERSHEY_SIMPLEX, 1.4, (255, 0, 0), 3)

    # Write and optionally display frame
    out.write(annotated)
    cv2.waitKey(1)

cap.release()
out.release()
cv2.destroyAllWindows()
```
